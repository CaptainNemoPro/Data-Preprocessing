```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
def count_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    return len([word for word in words if word.lower() in stop_words])
```

```python
group = df['Message'].apply(count_stopwords)
```

```python
df['count_stopwords'] = group
```

```python
df
```

```python
df.sort_values(by='count_stopwords', ascending=False).head(10)
```

```python
plt.bar(df['Message'], df['count_stopwords'])
plt.xlabel('Texts')
plt.ylabel('Stopword Counts')
plt.title('Stopword Counts by Text')
plt.show()
```
